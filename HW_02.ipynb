#### Исследовать функцию f(x) = x^3 - x^2 по плану:

#### Область задания и область значений

##### (-∞; ∞) 

#### Нули функции и их кратность

##### f(x) = x ^ 2 (x - 1) ===> x1 = 1, x2 = x3 = 0
##### x = 1 - имеет кратность 1
##### x = 0 - имеет кратность 2

```python
def f(x):
    return x ** 3 - x ** 2
```

```python
xs = np.linspace(-0.5, 1.5, 100)
```

```python
ys = f(xs)
```

```python
plt.figure(figsize=(16,9))
plt.plot(xs, ys, label = 'func 1')
plt.xscale('linear')
plt.yscale('linear')
plt.legend()
plt.grid()
plt.show()
```

```python
xs = np.linspace(-25, 25, 100)
ys = f(xs)
```

```python
plt.figure(figsize=(16,9))
plt.plot(xs, ys, label = 'func 1')
plt.legend()
plt.grid()
plt.show()
```

#### монотонность

##### f'(x) = 3 x ^ 2 - 2x = x ( 3x - 2)
##### f'(x) = 0 при x1 = 0, x2= 2/3

```python
def f_1(x):
    return x * (3 * x - 2)
```

```python
xs = np.linspace(-0.5, 1, 100)
ys = f_1(xs)
```

```python
plt.figure(figsize=(16,9))
plt.plot(xs, ys, label = 'derivative func')
plt.plot(xs, np.zeros_like(xs))
plt.plot([0, 2/3], [0, 0], 'ro')
plt.legend()
plt.grid()
plt.show()
```

##### функция f(x) = x^3 - x^2 возрастает на отрезках (-∞; 0) и (2/3; ∞)

##### функция нечетная, неограниченная, не является периодической
